# House Data

```python
HOUSE_CSV = '../raw_data/house_results/1976-2018-house.csv'
```

```python
# Read
house = pd.read_csv(HOUSE_CSV, delimiter=',', encoding='utf-8')
house.head()
```

```python
house = house.groupby(['state', 'state_po', 'year', 'party'])[['candidatevotes', 'totalvotes']].sum().reset_index()
```

```python
house['pop_percent'] = house['candidatevotes'] / house['totalvotes']
```

```python
house = house[['state', 'state_po', 'year', 'party', 'pop_percent']]
house.head()
```

```python
house = house[house.party.isin(['democrat', 'republican'])]
house.head()
```

```python
house.pivot_table(index=['state', 'year'], columns=['party'], values='pop_percent').fillna(0)
```
