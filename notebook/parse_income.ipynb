```python
import pandas as pd
import numpy as np
import datetime
import glob

UR_PATH = '../raw_data/personal_income'
SAVE_PATH= '../processed_data/personal_income.csv'
```

```python
# Read all income data for all states and concatenate
dfs = []
df = ''
for filename in glob.glob(UR_PATH + '/*.csv'):
    state_po = filename.split('/')[-1][:2]
    df = pd.read_csv(filename)
    df.rename(columns={state_po + 'PCPI': 'personal_income'}, inplace=True)
    df['state_po'] = state_po
    dfs.append(df)

all_inc = pd.concat(dfs)
all_inc
```

```python
all_inc['year'] = np.array([int(str(x).split('-')[0]) for x in all_inc['DATE']])
all_inc = all_inc.drop(columns=['DATE'])
```

```python
all_inc
```

```python
all_inc = all_inc.sort_values(by=['state_po', 'year'])
all_inc['percent_delta_prev'] = all_inc['personal_income'].pct_change()

all_inc = all_inc[(all_inc['year'] >= 1976) & (all_inc['year'] % 4 == 0)]
all_inc
```

```python
all_inc.to_csv(SAVE_PATH, index=False)
```
