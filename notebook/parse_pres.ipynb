```python
# Parse presidential election results

import pandas as pd
```

```python
PRES_CSV = '../raw_data/president_pop_vote_results/1976-2016-president.csv'
```

```python
df = pd.read_csv(PRES_CSV, delimiter=',', encoding='utf-8')
df.head()
```

```python
# Clean
df = df[df.party.isin(['democrat', 'republican'])]
df['pop_percent'] = df['candidatevotes'] / df['totalvotes']
df = df[['state', 'state_po', 'year', 'party', 'pop_percent']]
df.head(150)
```

```python
# Format
df = df.pivot_table(index=['year', 'state', 'state_po'], columns=['party'], values='pop_percent').reset_index()
df.head()
```
